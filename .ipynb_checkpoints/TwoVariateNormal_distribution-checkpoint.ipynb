# 二変量正規分布  

$$
    \begin{eqnarray}
    f(x, y) &=& \frac{1}{2 \pi \sqrt{\sigma_{x}^{2} \sigma_{y}^{2} (1 - \rho^2)}} \qquad (-\infty < x,y < \infty) \\
    q(x,y) &=& \frac{1}{1- \rho^2} \left[\left(\frac{x - \mu_x}{\sigma_x}\right)^2 - 2 \rho \left(\frac{x - \mu_x}{\sigma_x}\right)\left(\frac{y - \mu_y}{\sigma_y}\right)+\left(\frac{y - \mu_y}{\sigma_y}\right)^2 \right]
    \end{eqnarray}

```python
def twovar_normal(x_array, y_array):
    x_mu = np.mean(x_array)
    y_mu = np.mean(y_array)
    
    x_sigma = np.std(x_array)
    y_sigma = np.std(y_array)
    
    #ρ
    xy = np.array([x_array, y_array])
    low = np.cov(xy)[1,0] / np.sqrt((x_sigma ** 2) * (y_sigma ** 2))
    
    #[x1, y1], [x1, y2], [x1, y3], ... [xn, yn-1], [xn, yn]の順
    if (len(x_array) != len(y_array)):
        print("x shape and y shape mismatch")
    
    q = np.array([1 / (1 - (low ** 2)) * (((x_array[i] - x_mu) / x_sigma ) ** 2 - 2 * low * ((x_array[i] - x_mu) / x_sigma) * ((y_array[i] -y_mu) / y_sigma) + ((y_array[i] - y_mu) / y_sigma ) ** 2) \
        for i in range(len(x_array))])
    
    probability = np.array([1 / (2 * np.pi * np.sqrt((x_sigma**2) * (y_sigma**2) * (1 - low ** 2))) * np.exp((-qi) / 2) \
                  for qi in q])
    
    return probability
```

```python
x = stats.norm.rvs(loc=0, scale=1, size=1000)
y = stats.norm.rvs(loc=0, scale=1, size=1000)
prob = twovar_normal(x, y)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, prob, color='green')

plt.show()
```

```python
x = stats.norm.rvs(loc=0, scale=1, size=100)
y = stats.norm.rvs(loc=0, scale=1, size=100)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

Y, X = np.meshgrid(y, x)
prob = twovar_normal(X, Y)
ax.plot_wireframe(X, Y, prob, color='green')
plt.show()
```
