```python
import numpy as np
import math
from scipy.special import gamma, loggamma
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
```

# 二項分布

$$
\begin{eqnarray}
    P(X = x) \equiv f(x) &=& {}_n C_r p^x (1 - p)^{n - x}  (x = 0, 1, 2 , \dots, n)\\
                         &=& \frac{n!}{x!(n-x)!} p^x (1 - p)^{n - x}
\end{eqnarray}
$$


# 期待値

$$
E[X] = \sum_{x = 0}^{n} x p^x (1 - p)^{n - x} = n p
$$

# 分散

$$
    \begin{eqnarray}
    V[X] &=& E[X (X - 1)] + E[X] -{E[X]}^2 \\
         &=& n (n - 1) p^2 + np - (np)^2 \\
         &=& np(1 - p) \\
    \end{eqnarray}
$$

# グラフ

```python
def binomial(x_array, n, p):
    y = np.zeros(len(x_array)) 
    y = [math.factorial(n) / ((math.factorial(x_array[i])) * (math.factorial(n - x_array[i]))) * (p**x_array[1]) * ((1 - p)**(n - x_array[1])) for i in range(len(x_array))]
    return y
```

```python
x1 = np.arange(0,10,1)
y1 = binominal(x1, 10, 0.5)
x2 = np.arange(0, 20, 1)
y2 = binominal(x2, 20, 0.5)
x3 = np.arange(0, 30, 1)
y3 = binominal(x3, 30, 0.5)
x4 = np.arange(0, 50, 1)
y4 = binominal(x4, 50, 0.5)

plt.figure(figsize = (12, 8))
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.xlabel("numbers")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
p_array = np.arange(0.0, 1.01, 0.1)

fig = plt.figure(figsize=(12, 8))

def update(i):
    # 前フレームのグラフを初期化
    plt.cla()
    
    # i回目の値を取得
    p = p_array[i]
    
    probability = binomial(x1, 10, p)
    
    plt.bar(x1,probability,color = 'blue')
    plt.xlabel('x')
    plt.ylabel('probability') 
    plt.suptitle('Binomial Distribution', fontsize=20) 
    plt.title('p =' + str(np.round(p, 2)) + ', n =' + str(10), loc = 'left') # タイトル
#     plt.xticks(ticks=x_vals)
    plt.grid() # グリッド線
    plt.ylim(-0.1, 1.1) # y軸の表示範囲
```

```python
# gif画像を作成
anime_prob = FuncAnimation(fig, update, frames=len(p_array), interval=100)

# gif画像を保存
anime_prob.save('Binomial_prob_phi.gif')
```

## 期待値の計算

```python
#binomial distribution
def binomial_Ex(n,p):
    sum = 0
    
    #sigma calculation
    for x in range(n + 1):
        k = x * math.factorial(n) / math.factorial(x) / math.factorial(n - x) * (p ** x) * ((1-p) ** (n - x)) 
        sum += k
        print(f"{k}")
        if x < n:
            print(" + ")
        pass
    
    print(f"\n= {sum}")
```

```python
for i in range(5):
    n = random.randint(2,10)
    p =random.randint(2,10)
    print(f"binomial({n}, 1/{p})")
    binomial_Ex(n,1/p)
    print("\n")
```
