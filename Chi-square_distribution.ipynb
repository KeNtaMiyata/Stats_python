```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
```

```python
def chai_square(n):
    num = 10000
    sp.random.seed(0)
    x = np.array([stats.norm.rvs(size = n) for i in range(num)])
    x = np.power(x,2)
    w = np.array([np.sum(x[i]) for i in range(num)])
    return w
```

```python
w1 = chai_square(1)
w2 = chai_square(3)
w3 = chai_square(5)
w4 = chai_square(10)
w5 = chai_square(15)
```

```python
bins = np.linspace(-1,30,300)

plt.figure(figsize=(12,8))
plt.hist(w1, bins=bins, density = True, alpha = 0.5, label='n=1')
plt.hist(w2, bins=bins, density = True, alpha = 0.5, label='n=3')
plt.hist(w3, bins=bins, density = True, alpha = 0.5, label='n=5')
plt.hist(w4, bins=bins, density = True, alpha = 0.5, label='n=10')
plt.hist(w5, bins=bins, density = True, alpha = 0.5, label='n=15')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
