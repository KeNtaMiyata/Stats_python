```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# ガンマ関数
$$
    \Gamma(s) = \int_{0}^{\infty} x^{s-1} e^{-x} dx \qquad(s > 0)
$$

```python
def gammafunc(s):
    #input s →　output(s - 1)!
    dt = 0.001
    t = np.arange(dt, 100, dt)
    G = [(ti**(s-1)) * math.exp(-ti) * dt for ti in t]
    return np.sum(G)
```

$$
    \Gamma(\frac{1}{2}) = \sqrt(\pi) 
$$

```python
math.sqrt(np.pi)
```

```python
gammafunc(0.5)
```

```python
x = np.linspace(0.01, 5, 50)
y = [gammafunc(xi) for xi in x]

plt.figure(figsize = (12, 8))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
